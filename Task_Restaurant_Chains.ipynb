```python
# Import all the libraries we'll need

# for numeric calculations
import numpy as np
# for data handling
import pandas as pd
# for plotting charts
import matplotlib.pyplot as plt
# for making plots
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Datasets/Dataset .csv")
```

```python
# Take a quick look at the first 25 rows
df.head()
```

```python
df.tail()
```

```python
# Check how many rows and columns into the dataset (rows x columns)
df.shape
```

```python
# Count how many times each restaurant name appears
restaurant_counts = df['Restaurant Name'].value_counts()

# Filter only chains (more than 1 outlet)
chains = restaurant_counts[restaurant_counts > 1]


print("Total chains found:", len(chains))
chains.head(10)
```

```python
# Filter chain restaurants only
chain_df = df[df['Restaurant Name'].isin(chains.index)]
```

```python
# Group by Restaurant Name and get average rating & votes
chain_analysis = chain_df.groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',
    'Votes': 'mean',
    'Restaurant ID': 'count'
}).rename(columns={
    'Aggregate rating': 'Avg Rating',
    'Votes': 'Avg Votes',
    'Restaurant ID': 'Outlets'
}).sort_values(by='Outlets', ascending=False)

chain_analysis.head(10)
```

```python
# Look at top 10 restaurant chains by number of outlets

top_chains = chain_analysis.head(10)

# Plot bar chart for top 10 chains by outlets
plt.figure(figsize=(10,6))
sns.barplot(x=top_chains['Outlets'], y=top_chains.index, palette='mako')
plt.title("Top 10 Restaurant Chains by Number of Outlets")
plt.xlabel("Outlets")
plt.ylabel("Restaurant Chain")
plt.show()
```

```python
# Get top 10 chains based on average customer rating
top_rating_chains = chain_analysis.sort_values(by='Avg Rating', ascending=False).head(10)

# Plot bar chart for top rated chains
plt.figure(figsize=(10,6))
sns.barplot(x=top_rating_chains['Avg Rating'], y=top_rating_chains.index, palette='magma')
plt.title("Top 10 Chains by Avg Rating")
plt.xlabel("Average Rating")
plt.ylabel("Restaurant Chain")
plt.show()
```

###  Insights:

- The dataset contains several restaurant chains such as Barbeque Nation, Domino's Pizza, etc.
- Some chains have over 50+ outlets across cities.
- While some chains are highly popular (more votes), their average ratings vary.
- A few smaller chains show high ratings but fewer outlets.

###  Conclusion:

Restaurant chains dominate in number of outlets, but higher quantity doesn’t always mean higher quality. Rating and popularity both vary across brands.
